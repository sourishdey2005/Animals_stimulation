```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.animation import FuncAnimation
```

```python
def flower(u, v, t):
    # u: angle around (0 to 2Ï€)
    # v: radial distance (0 to 1)
    # t: time (animation frame)

    # Petal shape controlled by sinusoidal functions
    r = (1 + 0.3*np.sin(5*u + t)) * v

    # 3D coordinates
    x = r * np.cos(u) * (1 + 0.2*np.sin(t))
    y = r * np.sin(u) * (1 + 0.2*np.cos(t))
    z = 0.5*np.sin(3*v + t) + 0.2*np.cos(5*u + t)

    return x, y, z
```

```python
# Create meshgrid for flower surface
u = np.linspace(0, 2*np.pi, 100)
v = np.linspace(0, 1, 50)
U, V = np.meshgrid(u, v)

fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, projection='3d')

# Initial surface
X, Y, Z = flower(U, V, 0)
surf = ax.plot_surface(X, Y, Z, cmap='plasma', edgecolor='none')

def update(frame):
    ax.clear()
    X, Y, Z = flower(U, V, frame*0.1)
    surf = ax.plot_surface(X, Y, Z, cmap='hsv', edgecolor='none')

    # Set view and limits
    ax.set_xlim(-2,2)
    ax.set_ylim(-2,2)
    ax.set_zlim(-2,2)
    ax.set_axis_off()
    return surf,

ani = FuncAnimation(fig, update, frames=500, interval=50, blit=False)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.animation import FuncAnimation
```

```python
def flower(u, v, t):
    # u: angle around (0 to 2Ï€)
    # v: radial distance (0 to 1)
    # t: time (animation frame)

    # Petal radius changes with time
    r = (1 + 0.4*np.sin(6*u + t)) * v

    # 3D coordinates
    x = r * np.cos(u) * (1 + 0.2*np.sin(t))
    y = r * np.sin(u) * (1 + 0.2*np.cos(t))
    z = 0.5*np.sin(4*v + t) + 0.3*np.cos(6*u + t)

    return x, y, z
```

```python
# Create meshgrid for flower surface
u = np.linspace(0, 2*np.pi, 150)
v = np.linspace(0, 1, 80)
U, V = np.meshgrid(u, v)

fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, projection='3d')

# Initial surface
X, Y, Z = flower(U, V, 0)
surf = ax.plot_surface(X, Y, Z, cmap='hsv', edgecolor='none')

def update(frame):
    ax.clear()
    X, Y, Z = flower(U, V, frame*0.1)
    surf = ax.plot_surface(X, Y, Z, cmap='hsv', edgecolor='none')

    # Show coordinate axes
    ax.set_xlabel('X-axis')
    ax.set_ylabel('Y-axis')
    ax.set_zlabel('Z-axis')

    # Set limits
    ax.set_xlim(-2,2)
    ax.set_ylim(-2,2)
    ax.set_zlim(-2,2)

    # Rotate view for dynamic movement
    ax.view_init(elev=30, azim=frame)

    return surf,

ani = FuncAnimation(fig, update, frames=1000, interval=50, blit=False)
plt.show()
```
